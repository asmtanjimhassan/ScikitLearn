```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer
```

```python
df = pd.DataFrame({
    'sales': [10000, 3423, 45455, 56543],
    'city': ['Wisconsin', 'Atlanta', 'Wisconsin', 'New York'],
    'size': ['medium', 'small', 'large', 'large']
})

df
```

```python
ohe = OneHotEncoder(sparse_output=False)

sizes = ['small', 'medium', 'large']
ode = OrdinalEncoder(categories = [sizes])
```

```python
ct = make_column_transformer(
    (ohe, ['city']),
    (ode, ['size']),
    remainder = 'passthrough'
)
```

```python
ct.set_output(transform = 'pandas')
```

```python
df_pandas = ct.fit_transform(df)
```

```python
df_pandas
```

```python
ct2 = make_column_transformer(
    (ohe, ['city']),
    (ode, ['size']),
    remainder = 'drop'
)
```

```python
ct2.set_output(transform = 'pandas')
```

```python
df_pandas2 = ct2.fit_transform(df)
```

```python
df_pandas2
```

```python
ct3 = make_column_transformer(
    (ohe, ['city']),
    ('passthrough', ['size']),
    remainder = 'drop'
)
```

```python
ct3.set_output(transform = 'pandas')
```

```python
df_pandas3 = ct3.fit_transform(df)
```

```python
df_pandas3
```
