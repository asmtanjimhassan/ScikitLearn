```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.DataFrame({
    'farthest_run_mi' : [50, 62, np.NaN, 45, 45,65]
})
```

```python
df
```

```python
imp_mean = SimpleImputer(strategy = 'mean')

s1 = pd.DataFrame(imp_mean.fit_transform(df))
s1
```

```python
imp_median = SimpleImputer(strategy = 'median')

s2 = pd.DataFrame(imp_median.fit_transform(df))
s2
```

```python
imp_mod = SimpleImputer(strategy = 'most_frequent')
s3 = pd.DataFrame(imp_mod.fit_transform(df))

s3
```

```python
imp_cons = SimpleImputer(strategy = 'constant', fill_value = 56)
s4 = pd.DataFrame(imp_cons.fit_transform(df))
s4
```

```python
names = pd.DataFrame({
    'name' : ['Ryan', np.NaN, 'Pam', 'Jim', 'Michael', 'Stanley']
})
names
```

```python
imp_const_cat = SimpleImputer(strategy='constant', fill_value='NPC')
names1 = pd.DataFrame(imp_const_cat.fit_transform(names))

names1
```

```python
imp_mean_marked = SimpleImputer(strategy = 'mean', add_indicator=True)
imp_mean_marked.fit_transform(df)
```

```python
df2 = pd.concat([names, df], axis=1)
df2
```

```python
from sklearn.compose import make_column_transformer

ct = make_column_transformer(
    (imp_const_cat, ['name']),
    (imp_mean, ['farthest_run_mi']),
    remainder = 'drop'
)

ct.set_output(transform = 'pandas')

df3 = ct.fit_transform(df2)
df3
```
